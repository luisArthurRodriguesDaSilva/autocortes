```python
import os
def salvar_contexto(contexto,inicio,keywords:list[str],video_id):
    str_keywords = "_".join(keywords)
    caminho = f"../contextos/{video_id}_{inicio}_{str_keywords}.txt"
    os.makedirs(os.path.dirname(caminho), exist_ok=True)
    with open(caminho, "w", encoding="utf-8") as f:
        f.write(contexto)
```

```python
for keyworded_transcript_item in keyworded_trascript_items_without_redundant_appearances:
    contexto = fk.get_context(complete_trascript, keyworded_transcript_item, 30, 100)
    salvar_contexto(contexto,
                    keyworded_transcript_item["inicio"],
                    keyworded_transcript_item["keywords"],
                    video_id)
    
```
